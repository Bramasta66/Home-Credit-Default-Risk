### Explore Previous Application

**Explanation of Features in the Previous Application Dataset :**</br>
0. SK_ID_PREV </br>
ID of previous credit in Home credit related to loan in our sample. (One loan in our sample can have, or more previous loan applications in Home Credit, previous application could, but not necessarily have to lead to credit),hashed
1. SK_ID_CURR </br>
ID of loan in our sample,hashed </br>
2. NAME_CONTRACT_TYPE </br>
Contract product type (Cash loan, consumer loan [POS] ,...) of the previous application </br>
3. AMT_ANNUITY </br>
Annuity of previous application </br>
4. AMT_APPLICATION </br>
For how much credit did client ask on the previous application </br>
5. AMT_CREDIT </br>
Final credit amount on the previous application. This differs from AMT_APPLICATION in a way that the AMT_APPLICATION is the amount for which the client initially applied for, but during our approval process he could have received different amount - AMT_CREDIT </br>
6. AMT_DOWN_PAYMENT </br>
Down payment on the previous application </br>
7. AMT_GOODS_PRICE </br>
Goods price of good that client asked for (if applicable) on the previous application </br>
8. WEEKDAY_APPR_PROCESS_START </br>
On which day of the week did the client apply for previous application </br>
9. HOUR_APPR_PROCESS_START </br>
Approximately at what day hour did the client apply for the previous application,rounded </br>
10. FLAG_LAST_APPL_PER_CONTRACT </br>
Flag if it was last application for the previous contract. Sometimes by mistake of client or our clerk there could be more applications for one single contract </br>
11. NFLAG_LAST_APPL_IN_DAY </br>
Flag if the application was the last application per day of the client. Sometimes clients apply for more applications a day. Rarely it could also be error in our system that one application is in the database twice </br>
12. NFLAG_MICRO_CASH </br>
Flag Micro finance loan </br>
13. RATE_DOWN_PAYMENT </br>
Down payment rate normalized on previous credit,normalized </br>
14. RATE_INTEREST_PRIMARY </br>
Interest rate normalized on previous credit,normalized </br>
15. RATE_INTEREST_PRIVILEGED </br>
Interest rate normalized on previous credit,normalized </br>
16. NAME_CASH_LOAN_PURPOSE </br>
Purpose of the cash loan </br>
17. NAME_CONTRACT_STATUS </br>
Contract status (approved, cancelled, ...) of previous application </br>
18. DAYS_DECISION </br>
Relative to current application when was the decision about previous application made,time only relative to the application </br>
19. NAME_PAYMENT_TYPE </br>
Payment method that client chose to pay for the previous application </br>
20. CODE_REJECT_REASON </br>
Why was the previous application rejected </br>
21. NAME_TYPE_SUITE <
/br>
Who accompanied client when applying for the previous application </br>
22. NAME_CLIENT_TYPE </br>
Was the client old or new client when applying for the previous application </br>
23. NAME_GOODS_CATEGORY </br>
What kind of goods did the client apply for in the previous application </br>
24. NAME_PORTFOLIO </br>
Was the previous application for CASH, POS, CAR, â€¦, </br>
25. NAME_PRODUCT_TYPE </br>
Was the previous application x-sell o walk-in </br>
26. CHANNEL_TYPE </br>
Through which channel we acquired the client on the previous application </br>
27. SELLERPLACE_AREA </br>
Selling area of seller place of the previous application </br>
28. NAME_SELLER_INDUSTRY </br>
The industry of the seller </br>
29. CNT_PAYMENT </br>
Term of previous credit at application of the previous application </br>
30. NAME_YIELD_GROUP </br>
Grouped interest rate into small medium and high of the previous application,grouped </br>
31. PRODUCT_COMBINATION </br>
Detailed product combination of the previous application </br>
32. DAYS_FIRST_DRAWING </br>
Relative to application date of current application when was the first disbursement of the previous application,time only relative to the application </br>
33. DAYS_FIRST_DUE </br>
Relative to application date of current application when was the first due supposed to be of the previous application,time only relative to the application </br>
34. DAYS_LAST_DUE_ST_VERSION </br>
Relative to application date of current application when was the first due of the previous application,time only relative to the application </br>
35. DAYS_LAST_DUE </br>
Relative to application date of current application when was the last due date of the previous application,time only relative to the application </br>
36. DAYS_TERMINATION </br>
Relative to application date of current application when was the expected termination of the previous application,time only relative to the application </br>

**Statistical descriptive of all the feature**

```python
list_item_ap = []
for col in prev_app.columns:
    list_item_ap.append([
        col,
        prev_app[col].dtype,
        prev_app[col].count(),
        prev_app[col].isna().sum(),
        100*prev_app[col].isna().sum() / len(prev_app[col]),
        prev_app[col].nunique(),
        prev_app[col].unique()
    ])

desc_df = pd.DataFrame(data = list_item_ap, columns =[
    'feature', 'data_type', 'fill_num', 'null_num', 'null_persen', 'unique_num', 'unique_sample'
])

desc_df
```

**Check if there are any features with duplicate values**

```python
def check_duplicates(prev_app):
    data_duplikat = prev_app[prev_app.duplicated()]
    if not data_duplikat.empty:
        print(f"Ada {len(data_duplikat)} baris yang duplikat.")
        print("\nBaris yang duplikat:")
        print(data_duplikat)
    else:
        print("Tidak ada data yang duplikat di dataset.")
check_duplicates(prev_app)
```

**Check if there are any features with missing values:**

```python
def check_missing_values(prev_app):
    missing_values = prev_app.isnull().sum()
    if missing_values.any():
        print("Kolom yang terdapat missing value:")
        print(missing_values[missing_values > 0])
    else:
        print("Tidak ada missing value di dataset.")
check_missing_values(prev_app)
```

```python
missing_percentage = prev_app[['SELLERPLACE_AREA', 'DAYS_FIRST_DRAWING', 'DAYS_FIRST_DUE',
                               'DAYS_LAST_DUE_1ST_VERSION', 'DAYS_LAST_DUE', 'DAYS_TERMINATION']].isnull().mean() * 100
print(missing_percentage)
```

### Univariate Analysis

```python
# Identify numerical and categorical columns
numerical_columns = prev_app.select_dtypes(include=['int64', 'float64']).columns
categorical_columns = prev_app.select_dtypes(include=['object']).columns

print("Numerical Columns: ", numerical_columns)
print("Categorical Columns: ", categorical_columns)
```

**Numerical Analysis**

```python
# Boxplot untuk mengetahui berapa banyak dan berapa jauh outlier di setiap kolom
plt.figure(figsize=(16, 20))
for i, col in enumerate(numerical_columns):
    plt.subplot(len(numerical_columns)//2+1, 2, i+1)
    sns.boxplot(data=prev_app, x=col)
    plt.title(f'Boxplot of {col}')
    plt.tight_layout()
plt.show()
```

**Categorical Analysis**

```python
import matplotlib.pyplot as plt
import seaborn as sns

# List kolom kategorikal yang ingin dianalisis dari prev_app
categorical = ['NAME_CONTRACT_STATUS', 'NAME_CLIENT_TYPE', 'NAME_GOODS_CATEGORY','NAME_YIELD_GROUP', 'NAME_CASH_LOAN_PURPOSE', 'CODE_REJECT_REASON']

# Plot count plots untuk setiap kolom  # Mengatur ukuran plot
plt.figure(figsize=(16, 16))  
for i, col in enumerate(categorical):
    plt.subplot(3, 3, i+1)  # Membuat grid 3 baris dan 3 kolom
    sns.countplot(data=prev_app, x=col, order=prev_app[col].value_counts().index)
    plt.xticks(rotation=90) 
    plt.title(f'Distribution of {col}')
    plt.tight_layout()

plt.show()
```

### Multivariate Analysis

```python
# Import dataset yang dijadikan acuan, yaitu application_train
df_train = pd.read_csv('application_train.csv')

# Merge atau gabungkan dataset, tetapi hanya dengan kolom Target
merged_df = pd.merge(prev_app, df_train[['SK_ID_CURR', 'TARGET']], on='SK_ID_CURR', how='left')

merged_df.sample(10)
```

**Correlation Heatmap**

```python
# Kolom data numerik yang sudah dimerge
num_col_merge = merged_df.select_dtypes(include=['int64', 'float64']).columns

# Korelasi antar feature yang disertai feature target
correlation_matrix = merged_df[num_col_merge].corr()

# Heatmap figure
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap of prev_app Numerical Features')
plt.show()
```

```python
# Buat pivot table untuk mendapatkan jumlah masing-masing kategori
pivot_table = merged_df.pivot_table(index='NAME_CONTRACT_STATUS', columns='TARGET', aggfunc='size')

# Hitung jumlah total untuk setiap kategori dan urutkan
pivot_table['Total'] = pivot_table.sum(axis=1)
pivot_table_sorted = pivot_table.sort_values(by='Total', ascending=False)

# Buat stacked bar plot dengan urutan yang telah diatur
pivot_table_sorted.drop('Total', axis=1).plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Stacked Bar Plot of Contract Status by Target (Sorted by Total)')
plt.xticks(rotation=0)
plt.ylabel('Count')
plt.show()
```
