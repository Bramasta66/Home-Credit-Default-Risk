For how much credit did client ask on the previous application </br>

#### C. Handle Outliers

Handle Outliers menggunakan IQR karena data memiliki banyak outliers ekstrem. Handling outliers digunakan untuk data numerikal murni.
Pertama-tama definisikan terlebih dahulu kolom numerikal, kategorikal linear, kategorikal non-linear, dan kolom yang akan didelete untuk memudahkan proses selanjutnya.

```python
#data numerikal murni
num_col = prev_app[['AMT_ANNUITY', 'AMT_APPLICATION', 'AMT_CREDIT', 'AMT_DOWN_PAYMENT', 'AMT_GOODS_PRICE',
           'DAYS_DECISION', 'DAYS_FIRST_DRAWING', 'DAYS_FIRST_DUE', 'DAYS_LAST_DUE_1ST_VERSION', 'DAYS_LAST_DUE', 'DAYS_TERMINATION']]

#data kategori yang dibuat numerikal
#cats_lin = prev_app[['HOUR_APPR_PROCESS_START', 'NFLAG_LAST_APPL_IN_DAY', 'SELLERPLACE_AREA', 'CNT_PAYMENT', 'NFLAG_INSURED_ON_APPROVAL']]

#data kategori
#cats_nonlin = prev_app[['NAME_CONTRACT_TYPE', 'WEEKDAY_APPR_PROCESS_START', 'FLAG_LAST_APPL_PER_CONTRACT', 'NAME_CASH_LOAN_PURPOSE', 'NAME_CONTRACT_STATUS', 'NAME_PAYMENT_TYPE', 'CODE_REJECT_REASON','NAME_TYPE_SUITE', 'NAME_CLIENT_TYPE', 'NAME_GOODS_CATEGORY', 'NAME_PORTFOLIO', 'NAME_PRODUCT_TYPE','CHANNEL_TYPE', 'NAME_SELLER_INDUSTRY', 'NAME_YIELD_GROUP', 'PRODUCT_COMBINATION']]

#data kolom yang berupa normalized akan dihapus
#col_remove = prev_app[['RATE_DOWN_PAYMENT', 'RATE_INTEREST_PRIMARY', 'RATE_INTEREST_PRIVILEGED']]
```

```python
# Hitung Q1 dan Q3
Q1 = num_col.quantile(0.25)
Q3 = num_col.quantile(0.75)
IQR = Q3 - Q1

# Filter outliers menggunakan IQR
prev_app2 = prev_app[~((num_col < (Q1 - 1.5 * IQR)) | (num_col > (Q3 + 1.5 * IQR))).any(axis=1)]

print("Data setelah menghapus outliers menggunakan IQR:")
print(prev_app2.info())
```

```python
#Check kembali missing value setelah handling outlier
check_missing_values(prev_app2)
```

#### D. Feature Transformation

Sebelum transformasi data numerik, dilihat terlebih dahulu distribusi data untuk menentukan metode yang tepat.

```python
#Numerical features from latest data
num_col2 = prev_app2[['AMT_ANNUITY', 'AMT_APPLICATION', 'AMT_CREDIT', 'AMT_DOWN_PAYMENT', 'AMT_GOODS_PRICE',
           'DAYS_DECISION', 'DAYS_FIRST_DRAWING', 'DAYS_FIRST_DUE', 'DAYS_LAST_DUE_1ST_VERSION', 'DAYS_LAST_DUE', 'DAYS_TERMINATION']]

# Determine the distribution type of each numeric column
distribution_types = {}

for col in num_col2.columns:
    skewness = num_col2[col].skew()
    if skewness > 0.5:
        distribution_types[col] = 'Right-skewed'
    elif skewness < -0.5:
        distribution_types[col] = 'Left-skewed'
    else:
        distribution_types[col] = 'Normal'

# Display the results
for col, dist_type in distribution_types.items():
    print(f'{col}: {dist_type}')
```

```python
# Buat figure dan axes untuk 11 subplot
fig, axes = plt.subplots(nrows=4, ncols=3, figsize=(20, 15))  # 4 baris dan 3 kolom subplot
axes = axes.flatten()  # Ratakan array 2D menjadi 1D agar lebih mudah diakses

# Loop melalui setiap kolom dan plot histogram
for i, col in enumerate(num_col2):
    if i >= len(axes):  # Pastikan tidak melebihi jumlah axes yang tersedia
        break
    mean_val = num_col2[col].mean()
    median_val = num_col2[col].median()

    sns.histplot(num_col2[col], ax=axes[i], bins=20, kde=True, label=col)
    axes[i].axvline(x=mean_val, color='g', linestyle='dashed', label='Mean')
    axes[i].axvline(x=median_val, color='y', linestyle='dashed', label='Median')
    axes[i].legend()
    axes[i].set_title(f'Histogram of {col}')

# Hapus subplot yang tidak terpakai (jika ada)
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

# Atur layout agar tidak saling tumpang tindih
plt.tight_layout()
plt.show()
```

```python
#Lakukan transformasi dengan MinMaxScaler pada data yang Left-skewed
from sklearn.preprocessing import MinMaxScaler

prev_app2.loc[:, 'DAYS_DECISION'] = MinMaxScaler().fit_transform(prev_app2['DAYS_DECISION'].values.reshape(len(prev_app2), 1))
prev_app2.loc[:, 'DAYS_TERMINATION'] = MinMaxScaler().fit_transform(prev_app2['DAYS_FIRST_DRAWING'].values.reshape(len(prev_app2), 1))
```

```python
#Lakukan strandardized pada data yang berdistribusi Normal
from sklearn.preprocessing import StandardScaler
prev_app2.loc[:, 'AMT_DOWN_PAYMENT'] = StandardScaler().fit_transform(prev_app2['AMT_DOWN_PAYMENT'].values.reshape(len(prev_app2), 1))
prev_app2.loc[:, 'DAYS_FIRST_DRAWING'] = StandardScaler().fit_transform(prev_app2['DAYS_FIRST_DRAWING'].values.reshape(len(prev_app2), 1))
prev_app2.loc[:, 'DAYS_FIRST_DUE'] = StandardScaler().fit_transform(prev_app2['DAYS_FIRST_DUE'].values.reshape(len(prev_app2), 1))
prev_app2.loc[:, 'DAYS_LAST_DUE_1ST_VERSION'] = StandardScaler().fit_transform(prev_app2['DAYS_LAST_DUE_1ST_VERSION'].values.reshape(len(prev_app2), 1))
prev_app2.loc[:, 'DAYS_LAST_DUE'] = StandardScaler().fit_transform(prev_app2['DAYS_LAST_DUE'].values.reshape(len(prev_app2), 1))
```

```python
#Lakukan transformasi dengan Log Transformation pada data yang Right-skewed
#np.log1p() calculates the natural logarithm of (1 + x), which can handle zero values:
prev_app2.loc[:, 'AMT_ANNUITY'] = np.log1p(prev_app2['AMT_ANNUITY'])
prev_app2.loc[:, 'AMT_APPLICATION'] = np.log1p(prev_app2['AMT_APPLICATION'])
prev_app2.loc[:, 'AMT_CREDIT'] = np.log1p(prev_app2['AMT_CREDIT'])
prev_app2.loc[:, 'AMT_GOODS_PRICE'] = np.log1p(prev_app2['AMT_GOODS_PRICE'])
```
