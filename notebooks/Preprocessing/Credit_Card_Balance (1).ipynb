### HANDLING OUTLIERS

```python
print(f'Jumlah baris sebelum memfilter outlier: {len(merged_df)}')

filtered_entries = np.array([True] * len(merged_df))
for col in ['MONTHS_BALANCE', 'AMT_BALANCE', 'AMT_CREDIT_LIMIT_ACTUAL', 'AMT_DRAWINGS_ATM_CURRENT', 'AMT_DRAWINGS_CURRENT', 'AMT_DRAWINGS_OTHER_CURRENT', 'AMT_DRAWINGS_POS_CURRENT', 'AMT_INST_MIN_REGULARITY', 'AMT_PAYMENT_CURRENT', 'AMT_PAYMENT_TOTAL_CURRENT', 'AMT_RECEIVABLE_PRINCIPAL', 'AMT_RECIVABLE', 'AMT_TOTAL_RECEIVABLE', 'CNT_DRAWINGS_ATM_CURRENT', 'CNT_DRAWINGS_CURRENT', 'CNT_DRAWINGS_OTHER_CURRENT', 'CNT_DRAWINGS_POS_CURRENT', 'CNT_INSTALMENT_MATURE_CUM', 'SK_DPD', 'SK_DPD_DEF']:
    Q1 = merged_df[col].quantile(0.25)
    Q3 = merged_df[col].quantile(0.75)
    IQR = Q3 - Q1
    low_limit = Q1 - (IQR * 1.5)
    high_limit = Q3 + (IQR * 1.5)

    filtered_entries = ((merged_df[col] >= low_limit) & (merged_df[col] <= high_limit)) & filtered_entries

merged_df = merged_df[filtered_entries]

print(f'Jumlah baris setelah memfilter outlier: {len(merged_df)}')
```

Outliers adalah nilai yang sangat berjauhan dengan data umumnya. Ini dapat mempengaruhi kinerja suatu algoritma atau model. Terdapat jumlah outliers pada dataset Credit_Card_Balance sebesar **3227965**. Hal ini perlu dilakukan handling untuk mendapatkan data yang rapi agar nantinya model dapat menghasilkan kinerja yang sangat baik. Jika tidak dilakukan penanganan ini maka dapat meningkatkan pengaruh negatif pada algoritma yang sensitif terhadp outliers. Setelah dilakukan handling outliers, maka didapatkan Jumlah baris setelah memfilter outlier sebesar **2213965**

### FEATURE ENGINEERING

Feature Engineering diperlukan untuk menambah kompleksitas data agar pada saat pemodelan didapatkan hasil dengan kinerja yang paling baik. Hal-hal yang dilakukan adalah melakukan penambahan feature baru dari feature-feature yang sudah ada. Berikut feature-feature yang perlu ditambahkan:

**Payment to Balance Ratio**  : Rasio pembayaran terhadap saldo yang ada, yang menunjukkan seberapa baik peminjam mengelola hutangnya. Contoh: AMT_PAYMENT_TOTAL_CURRENT / AMT_BALANCE.

**Utilization Rate**          : Rasio antara jumlah kredit yang digunakan terhadap batas kredit. Contoh: AMT_BALANCE / AMT_CREDIT_LIMIT_ACTUAL.

**Regularization**          : Rasio antara jumlah yang harus dibayar terhadap umlah cicilan atau angsuran minimal yang dibayar secara berkala dalam jangka waktu tertentu. Contoh: AMT_PAYMENT_CURRENT / AMT_INST_MIN_REGULARITY.




```python
#Menambah Feature Baru
merged_df['UTILIZATION_RATE'] = merged_df['AMT_BALANCE'] / merged_df['AMT_CREDIT_LIMIT_ACTUAL']
merged_df['PAYMENT_TO_BALANCE_RATIO'] = merged_df['AMT_PAYMENT_TOTAL_CURRENT'] / merged_df['AMT_BALANCE']
merged_df['Regularization'] = merged_df['AMT_PAYMENT_CURRENT'] / merged_df['AMT_INST_MIN_REGULARITY']
```
